```python
import pandas as pd
import json

def jsonl_to_csv_pandas(input_file, output_file):
    data = []
    with open(input_file, 'r') as f:
        for line in f:
            data.append(json.loads(line.strip()))
    df = pd.json_normalize(data)  # Handles nesting automatically
    df.to_csv(output_file, index=False)

# Usage
jsonl_to_csv_pandas('media.jsonl', 'output.csv')
```
